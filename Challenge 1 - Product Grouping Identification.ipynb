```python
import numpy as np
import pandas as pd
import re
```

```python
em2 = pd.read_csv("data/Extra Material 2 - keyword list_with substring.csv")
em3 = pd.read_csv("data/Extra Material 3 - mismatch list.csv")
kw = pd.read_csv("data/Keyword_spam_question.csv")
```

```python
"""
Remove regex from keyword search
"""
```

```python
# Split function is required to assist in identifying mismatched words
# Without splitting, mismatch keywords that are a subset words in the sentence will be removed by mistake
# E.g. table appearing in inflatable will remove the word inflatable as well
```

```python
kw['mod_string'] = kw['name'].apply(lambda x: re.sub(r'[^\w ]', '', x).lower().split())
```

```python
kw['mod_string'].head(5)
```

```python
"""
Combining mismatch words for filtering
"""
```

```python
em3_transposed = em3.T.replace(np.NaN,'')
em3_transposed['words'] = list(zip(em3_transposed[0],em3_transposed[1],em3_transposed[2],em3_transposed[3]))
```

```python
em3_transposed.head(5)
```

```python
"""
Filtering mismatch words from keywords
"""
```

```python
# Making a copy of the modified string
```

```python
# Defining filter function, remove mismatched words only if they appear in pairs
# Flag is needed to ensure mismatched keyword is removed only if its corresponding pair appears in the sentence
```

```python
def filtering_func(sentence,words,index):
    for i in words:
        if i in sentence:
            sentence.remove(i)
            if index in sentence:
                sentence.remove(index)
        else:
            continue
    return sentence
```

```python
for i in em3_transposed.index:
    for j in range(len(kw)):
        if i in kw['mod_string'].iloc[j]:
            filtering_func(kw['mod_string'].iloc[j],em3_transposed['words'].loc[i],i)
```

```python
kw.head(5)
```

```python
"""
Removing regex from keywords categories
"""
```

```python
em2['mod_Keywords'] = em2['Keywords'].apply(lambda x: re.sub(r'[^\w ,]', '', x).lower().split(', '))
```

```python
"""
Evaluating product categories from filtered results
"""
```

```python
# Comparison of product categories with keyword search
```

```python
kw['evaluate'] = kw['mod_string'].apply(lambda x: ' '.join(x))
kw['groupings'] = ''
kw['keywords'] = ''
for i in range(len(kw.head(1000))):
    for j in range(len(em2)):
        for k in em2['mod_Keywords'].iloc[j]:
            if k in kw['evaluate'].iloc[i]:
            
            # If keyword appears in more than two product groupings, take the first instance
                if k not in kw['keywords'].iloc[i]:
                    kw.at[i,'keywords'] += (str(k) + ',')
                    kw.at[i,'groupings'] += (str(j) + ',')
                        
kw['groupings'] = kw['groupings'].apply(lambda x: x.split(',')[:-1])
kw['keywords'] = kw['keywords'].apply(lambda x: x.split(',')[:-1])
```

```python
kw.head(1000)
```

```python
"""
For the product groupings, if any are a substring of subsequent groupings, 
remove the subset keyword and product category as the longer word has priority 
"""
```

```python
# Compare each keyword with subsequent keywords in the list
# If the keyword is a subset, remove the subset

# Weird error in which value k sometimes has the tendency to exceed list range, resulting in an IndexError
```

```python
for i in range(len(kw.head(1000))):
        for j in range(len(kw['keywords'].iloc[i])):
            for k in range((j+1),len(kw['keywords'].iloc[i])):
                try:
                    if k == len(kw['keywords'].iloc[i]):
                        continue
                    else:
                        if kw['keywords'].iloc[i][j] in kw['keywords'].iloc[i][k]:
                            kw['keywords'].iloc[i].pop(j)
                            kw['groupings'].iloc[i].pop(j)
                except IndexError:
                    print(i,j,k)
```

```python
# Remove duplicate values while maintaining order
```

```python
def removeDuplicates(itemList):
    uniquelist = []
    for elem in itemList:
        if elem not in uniquelist:
            uniquelist.append(elem)
    return uniquelist
```

```python
kw['groups_found'] = kw['groupings'].apply(lambda x: removeDuplicates(x))
```

```python
kw.head(1000)
```

```python
"""
Output the required results into a csv file
"""
```

```python
results = kw.filter(['index','groups_found'], axis = 1)
```
